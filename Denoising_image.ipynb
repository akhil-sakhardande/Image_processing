# Python code for denoising image

```python
# %%writefile Denoising_image.ipynb

# !/usr/bin/python
# Created: Oct 7 2016
# Author: Akhil Sakhardande

import numpy as np
import scipy
#import scipy.misc
from scipy import ndimage
import matplotlib.pyplot as plt
 
import cv2
 
image = cv2.imread('/Users/akhilsakhardande/Documents/Work/KinRep/img.tiff')

#image = scipy.misc.face(gray=True)
#image = image[230:290, 220:320]
 
#noisy = image + 0.4*image.std()*np.random.random(image.shape)
 
gauss_denoised = ndimage.gaussian_filter(image, 2)
med_denoised = ndimage.median_filter(image, 3)
 
plt.figure(figsize=(12,2.8))
 
plt.subplot(131)
plt.imshow(image, cmap=plt.cm.gray, vmin=40, vmax=220)
plt.axis('off')
plt.title('original', fontsize=20)
plt.subplot(132)
 
plt.imshow(gauss_denoised, cmap=plt.cm.gray, vmin=40, vmax=220)
plt.axis('off')
plt.title('Gaussian filter', fontsize=20)
plt.subplot(133)
plt.imshow(med_denoised, cmap=plt.cm.gray, vmin=40, vmax=220)
plt.axis('off')
plt.title('Median filter', fontsize=20)
 
plt.subplots_adjust(wspace=0.02, hspace=0.02, top=0.9, bottom=0, left=0,
                    right=1)
plt.show()
 
cv2.imshow('image',image) # expects true color
cv2.waitKey(0)
 
 
cv2.imshow('image',gauss_denoised) # expects true color
cv2.waitKey(0)
 
cv2.imshow('image',med_denoised) # expects true color
cv2.waitKey(0)
 
```
